```python
def cesar_Code_mot(mot: str , decalage: int) -> str:
    mot_code = ""
    for lettre in mot:
        if lettre.isalpha():
            code_lettre = (ord(lettre) - ord('a') + decalage) % 26 + ord('a')
            mot_code += chr(code_lettre)
        else:
            mot_code += lettre
    return mot_code

print(cesar_Code_mot("bonjour", 3))

def cesar_Decode_mot(mot_code: str, decalage: int) -> str:
    mot_decode = ""
    for lettre in mot_code:
        if lettre.isalpha():
            code_lettre = (ord(lettre) - ord('a') - decalage) % 26 + ord('a')
            mot_decode += chr(code_lettre)
        else:
            mot_decode += lettre
    return mot_decode

print(cesar_Decode_mot("erqmrxu", 3))
```

```python
def coincidence_indice(text: str) -> float:
    freq = [0] * 26
    total_letters = 0

    for char in text:
        if char.isalpha():
            index = ord(char.lower()) - ord("a")
            freq[index] += 1
            total_letters += 1

    if total_letters == 0:
        return 0.0

    ic = sum(f * (f - 1) for f in freq) / (total_letters * (total_letters - 1))
    return ic


print(
    coincidence_indice(
        """hwfvsfl ds kwugfvw ymwjjw egfvasdw dwk sewjausafk wehdgqwjwfl 
vwk afvawfk fsnsbgk hgmj ujqhlwj vwk ewkksywk. u'wkl d'mf vwk jsjwk ugvwk vw d'zaklgajw s 
f'sngaj bsesak wlw tjakw. d'aehwfwljstadalw vm ugvw fsnsbg nawfl wf hsjlaumdawj vm xsal 
imw uwllw dsfymw f's smumf dawf snwu mfw imwdugfimw dsfymw wmjghwwffw gm 
skaslaimw."""
    )
)


print(
    coincidence_indice(
        """Ghpdlq, ghv o'dxeh, d o'khxuh rx eodqfklw od fdpsdjqh, Mh 
sduwludl. Yrlv-wx, mh vdlv txh wx p'dwwhqgv. M'ludl sdu od iruhw, m'ludl sdu od prqwdjqh. 
Mh qh sxlv ghphxuhu orlq gh wrl soxv orqjwhpsv. Mh pdufkhudl ohv bhxa ilahv vxu phv 
shqvhhv, Vdqv ulhq yrlu dx ghkruv, vdqv hqwhqguh dxfxq euxlw, Vhxo, lqfrqqx, oh grv frxueh,
ohv pdlqv furlvhhv, Wulvwh, hw oh mrxu srxu prl vhud frpph od qxlw. Mh qh uhjdughudl ql 
o'ru gx vrlu txl wrpeh, Ql ohv yrlohv dx orlq ghvfhqgdqw yhuv Kduiohxu, Hw txdqg 
m'duulyhudl, mh phwwudl vxu wd wrpeh Xq erxtxhw gh krxa yhuw hw gh euxbhuh hq iohxu. 
Ylfwru Kxjr - Ohv Frqwhpsodwlrqv"""
    )
)
```

```python
frenchLetterFreq = {
    'E': 14.72, 'A': 7.64, 'S': 7.95, 'I': 7.53, 'T': 7.24, 'N': 7.10, 'R': 6.55,
    'U': 6.31, 'L': 5.46, 'O': 5.27, 'D': 3.67, 'C': 3.26, 'P': 3.01, 'M': 2.97,
    'V': 1.63, 'Q': 1.36, 'F': 1.07, 'B': 0.90, 'G': 0.87, 'H': 0.74, 'J': 0.55,
    'X': 0.39, 'Y': 0.31, 'Z': 0.14, 'K': 0.05, 'W': 0.05
}

englishLetterFreq = {
    'E': 12.70, 'T': 9.06, 'A': 8.17, 'O': 7.51, 'I': 6.97, 'N': 6.75, 'S': 6.33,
    'H': 6.09, 'R': 5.99, 'D': 4.25, 'L': 4.03, 'C': 2.78, 'U': 2.76, 'M': 2.41,
    'W': 2.36, 'F': 2.23, 'G': 2.02, 'Y': 1.97, 'P': 1.93, 'B': 1.29, 'V': 0.98,
    'K': 0.77, 'J': 0.15, 'X': 0.15, 'Q': 0.10, 'Z': 0.07
}

ETAOIN_FR = 'EASITNRULODCPMVQFBGHJXYZWK'
ETAOIN_EN = 'ETAOINSHRDLCUMWFGYPBVKJXQZ'
LETTERS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'


def calculer_frequence_lettres(text: str) -> dict:
    """Calcule la fréquence de chaque lettre dans un texte."""
    freq = {}
    total = 0
    
    for char in text.upper():
        if char.isalpha():
            freq[char] = freq.get(char, 0) + 1
            total += 1
    
    for letter in freq:
        freq[letter] = (freq[letter] / total) * 100 if total > 0 else 0
    
    return freq


def trier_par_frequence(freq: dict) -> str:
    """Retourne les lettres triées par fréquence décroissante."""
    return ''.join(sorted(freq.keys(), key=lambda x: freq[x], reverse=True))


def trouver_cle_cesar_par_frequence(text: str, langue: str = "Français") -> int:
    """
    Trouve la clé César en comparant la lettre la plus fréquente du texte
    avec la lettre la plus fréquente de la langue.
    """
    freq = calculer_frequence_lettres(text)
    lettres_triees = trier_par_frequence(freq)
    
    if not lettres_triees:
        return 0
    
    lettre_plus_freq_chiffre = lettres_triees[0]
    
    if langue == "Français":
        lettre_plus_freq_langue = 'E'
    else:
        lettre_plus_freq_langue = 'E'
    
    cle = (ord(lettre_plus_freq_chiffre) - ord(lettre_plus_freq_langue)) % 26
    
    return cle


def dechiffrer_par_analyse_frequentielle(text: str) -> tuple[str, int, str]:
    """
    Déchiffre un message chiffré par César en utilisant l'analyse fréquentielle.
    Utilise d'abord l'indice de coïncidence pour déterminer la langue.
    Retourne (texte_déchiffré, clé, langue)
    """
    langue, ic = detect_langue(text)
    
    if abs(ic - 0.0778) < 0.015:
        langue = "Français"
    
    freq = calculer_frequence_lettres(text)
    lettres_triees = trier_par_frequence(freq)
    
    print(f"Analyse fréquentielle du texte:")
    print(f"  Indice de coïncidence: {ic:.4f}")
    print(f"  Langue probable: {langue}")
    print(f"  Lettres par fréquence: {lettres_triees[:10]}...")
    print(f"  Top 5 fréquences: ", end="")
    for i, letter in enumerate(lettres_triees[:5]):
        print(f"{letter}={freq[letter]:.1f}%", end=" ")
    print()
    
    cle = trouver_cle_cesar_par_frequence(text, langue)
    
    texte_dechiffre = cesar_Decode_maj_min(text, cle)
    
    return texte_dechiffre, cle, langue


print("=" * 70)
print("EXERCICE 3.1 - Déchiffrement par analyse fréquentielle")
print("=" * 70)

print("\n" + "-" * 70)
print("MESSAGE 4:")
print("-" * 70)
resultat4, cle4, langue4 = dechiffrer_par_analyse_frequentielle(message4)
print(f"\n  Clé trouvée: {cle4}")
print(f"\n  Texte déchiffré:\n{resultat4}")

print("\n" + "-" * 70)
print("MESSAGE 5:")
print("-" * 70)
resultat5, cle5, langue5 = dechiffrer_par_analyse_frequentielle(message5)
print(f"\n  Clé trouvée: {cle5}")
print(f"\n  Texte déchiffré:\n{resultat5}")
```

### 3.2 - Analyse par digrammes et trigrammes

```python
DIGRAMMES_FR = ['ES', 'DE', 'LE', 'EN', 'RE', 'NT', 'ON', 'ER', 'TE', 'EL',
                'AN', 'SE', 'ET', 'LA', 'AI', 'IT', 'ME', 'OU', 'EM', 'IE']

TRIGRAMMES_FR = ['ENT', 'LES', 'EDE', 'DES', 'QUE', 'AIT', 'LLE', 'SDE', 'ION', 'EME',
                 'ELA', 'RES', 'MEN', 'ESE', 'DEL', 'ANT', 'TIO', 'PAR', 'ESD', 'TDE']

DIGRAMMES_EN = ['TH', 'HE', 'IN', 'ER', 'AN', 'RE', 'ON', 'AT', 'EN', 'ND',
                'TI', 'ES', 'OR', 'TE', 'OF', 'ED', 'IS', 'IT', 'AL', 'AR']

TRIGRAMMES_EN = ['THE', 'AND', 'ING', 'ION', 'TIO', 'ENT', 'ERE', 'HER', 'ATE', 'VER',
                 'TER', 'THA', 'HAT', 'ERS', 'HIS', 'RES', 'ILL', 'ARE', 'CON', 'NCE']


def compter_digrammes(text: str) -> dict:
    """
    Compte la fréquence d'apparition de tous les digrammes (groupes de 2 lettres) dans un texte.
    Retourne un dictionnaire {digramme: count}
    """
    texte_propre = ''.join(c.upper() for c in text if c.isalpha())
    
    digrammes = {}
    for i in range(len(texte_propre) - 1):
        digramme = texte_propre[i:i+2]
        digrammes[digramme] = digrammes.get(digramme, 0) + 1
    
    return digrammes


def compter_trigrammes(text: str) -> dict:
    """
    Compte la fréquence d'apparition de tous les trigrammes (groupes de 3 lettres) dans un texte.
    Retourne un dictionnaire {trigramme: count}
    """
    texte_propre = ''.join(c.upper() for c in text if c.isalpha())
    
    trigrammes = {}
    for i in range(len(texte_propre) - 2):
        trigramme = texte_propre[i:i+3]
        trigrammes[trigramme] = trigrammes.get(trigramme, 0) + 1
    
    return trigrammes


def trier_ngrammes(ngrammes: dict, top_n: int = 10) -> list:
    """Retourne les n-grammes triés par fréquence décroissante."""
    return sorted(ngrammes.items(), key=lambda x: x[1], reverse=True)[:top_n]


def calculer_score_correspondance(ngrammes_texte: dict, ngrammes_ref: list) -> float:
    """
    Calcule un score de correspondance entre les n-grammes d'un texte
    et une liste de référence de n-grammes fréquents.
    Plus le score est élevé, plus le texte correspond à la langue de référence.
    """
    total = sum(ngrammes_texte.values())
    if total == 0:
        return 0.0
    
    score = 0
    for ngramme in ngrammes_ref:
        if ngramme in ngrammes_texte:
            poids = len(ngrammes_ref) - ngrammes_ref.index(ngramme)
            score += (ngrammes_texte[ngramme] / total) * poids
    
    return score


def est_texte_clair(text: str, seuil_digramme: float = 0.01, seuil_trigramme: float = 0.005) -> tuple[bool, str, dict]:
    """
    Détermine si un texte est en clair (non chiffré) en analysant les digrammes et trigrammes.
    
    Retourne:
    - bool: True si le texte semble être en clair
    - str: La langue détectée ("Français", "Anglais", ou "Indéterminé")
    - dict: Dictionnaire avec les scores et détails de l'analyse
    """
    digrammes = compter_digrammes(text)
    trigrammes = compter_trigrammes(text)
    
    score_di_fr = calculer_score_correspondance(digrammes, DIGRAMMES_FR)
    score_di_en = calculer_score_correspondance(digrammes, DIGRAMMES_EN)
    score_tri_fr = calculer_score_correspondance(trigrammes, TRIGRAMMES_FR)
    score_tri_en = calculer_score_correspondance(trigrammes, TRIGRAMMES_EN)
    
    score_fr = score_di_fr + score_tri_fr
    score_en = score_di_en + score_tri_en
    
    details = {
        'score_digrammes_fr': score_di_fr,
        'score_digrammes_en': score_di_en,
        'score_trigrammes_fr': score_tri_fr,
        'score_trigrammes_en': score_tri_en,
        'score_total_fr': score_fr,
        'score_total_en': score_en,
        'top_digrammes': trier_ngrammes(digrammes, 10),
        'top_trigrammes': trier_ngrammes(trigrammes, 10)
    }
    
    max_score = max(score_fr, score_en)
    
    if max_score > seuil_digramme + seuil_trigramme:
        est_clair = True
        langue = "Français" if score_fr > score_en else "Anglais"
    else:
        est_clair = False
        langue = "Indéterminé (probablement chiffré)"
    
    return est_clair, langue, details


def analyser_message_ngrammes(text: str, nom_message: str = ""):
    """Analyse complète d'un message avec affichage formaté."""
    print(f"\n{'=' * 70}")
    print(f"ANALYSE PAR DIGRAMMES/TRIGRAMMES - {nom_message}")
    print('=' * 70)
    
    est_clair, langue, details = est_texte_clair(text)
    
    print(f"\n📊 Top 10 digrammes trouvés:")
    for di, count in details['top_digrammes']:
        marqueur = "✓" if di in DIGRAMMES_FR or di in DIGRAMMES_EN else " "
        print(f"   {marqueur} {di}: {count}")
    
    print(f"\n📊 Top 10 trigrammes trouvés:")
    for tri, count in details['top_trigrammes']:
        marqueur = "✓" if tri in TRIGRAMMES_FR or tri in TRIGRAMMES_EN else " "
        print(f"   {marqueur} {tri}: {count}")
    
    print(f"\n📈 Scores de correspondance:")
    print(f"   Français  - Digrammes: {details['score_digrammes_fr']:.4f}, Trigrammes: {details['score_trigrammes_fr']:.4f}")
    print(f"   Anglais   - Digrammes: {details['score_digrammes_en']:.4f}, Trigrammes: {details['score_trigrammes_en']:.4f}")
    print(f"   Total FR: {details['score_total_fr']:.4f} | Total EN: {details['score_total_en']:.4f}")
    
    print(f"\n🔍 VERDICT:")
    if est_clair:
        print(f"   ✅ Le message semble être EN CLAIR ({langue})")
    else:
        print(f"   🔒 Le message semble être CHIFFRÉ/CODÉ")
    
    return est_clair, langue, details


# Test avec les messages déchiffrés vs chiffrés
print("=" * 70)
print("COMPARAISON: Messages chiffrés vs déchiffrés")
print("=" * 70)

# Message 4 chiffré
analyser_message_ngrammes(message4, "Message 4 (CHIFFRÉ)")

# Message 4 déchiffré
analyser_message_ngrammes(resultat4, "Message 4 (DÉCHIFFRÉ)")

# Message 6 chiffré
analyser_message_ngrammes(message6, "Message 6 (CHIFFRÉ)")

# Message 6 déchiffré
analyser_message_ngrammes(resultat6, "Message 6 (DÉCHIFFRÉ)")
```
