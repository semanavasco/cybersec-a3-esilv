## Exercice 1 - Code César
#### 1.1

```python
def caesar_encode(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

def caesar_decode(text, shift):
    return caesar_encode(text, -shift)

encoded = caesar_encode("cybersecurite", 11)
print("Chiffrement de 'cybersecurite' avec clé 11 :", encoded)
print(f"Déchiffrement de '{encoded}' avec la clé 11 :", caesar_decode(encoded, 11))
```

```python
def brute_force_caesar(cipher_text):
    for key in range(26):
        print(f"Clé {key:2d}: {caesar_decode(cipher_text, key)}")

message1 = "WP NZOLRP PDE FY LCE"
message2 = "HA YKZWCA AOP QJ WNP"

print("============ Message 1 ============")
brute_force_caesar(message1)

print("\n============ Message 2 ============")
brute_force_caesar(message2)
```

## Exercice 2 – Indice de Coïncidence

```python
def coincidence_index(text):
    freq = [0] * 26
    total_letters = 0
    for char in text:
        if char.isalpha():
            freq[ord(char.lower()) - ord('a')] += 1
            total_letters += 1
    if total_letters <= 1:
        return 0.0
    return sum(f * (f - 1) for f in freq) / (total_letters * (total_letters - 1))

INDICES = {
    "Suédois": 0.0644, "Serbe": 0.0643, "Russe": 0.0529, "Portugais": 0.0745,
    "Néerlandais": 0.0798, "Norvégien": 0.0694, "Malaysien": 0.0852,
    "Japonais": 0.0772, "Italien": 0.0738, "Hébreu": 0.0768, "Grec": 0.0691,
    "Français": 0.0778, "Finnois": 0.0737, "Esperanto": 0.069, "Espagnol": 0.077,
    "Danois": 0.0707, "Arabe": 0.0758, "Anglais": 0.0667, "Allemand": 0.0762,
}

def detect_language(text):
    ic = coincidence_index(text)
    closest = min(INDICES, key=lambda lang: abs(INDICES[lang] - ic))
    return closest, ic
```

```python
FRENCH_FREQ = {
    'E': 14.72, 'A': 7.64, 'S': 7.95, 'I': 7.53, 'T': 7.24, 'N': 7.10, 'R': 6.55,
    'U': 6.31, 'L': 5.46, 'O': 5.27, 'D': 3.67, 'C': 3.26, 'P': 3.01, 'M': 2.97,
    'V': 1.63, 'Q': 1.36, 'F': 1.07, 'B': 0.90, 'G': 0.87, 'H': 0.74, 'J': 0.55,
    'X': 0.39, 'Y': 0.31, 'Z': 0.14, 'K': 0.05, 'W': 0.05
}

ENGLISH_FREQ = {
    'E': 12.70, 'T': 9.06, 'A': 8.17, 'O': 7.51, 'I': 6.97, 'N': 6.75, 'S': 6.33,
    'H': 6.09, 'R': 5.99, 'D': 4.25, 'L': 4.03, 'C': 2.78, 'U': 2.76, 'M': 2.41,
    'W': 2.36, 'F': 2.23, 'G': 2.02, 'Y': 1.97, 'P': 1.93, 'B': 1.29, 'V': 0.98,
    'K': 0.77, 'J': 0.15, 'X': 0.15, 'Q': 0.10, 'Z': 0.07
}

ETAOIN_FR = 'EASITNRULODCPMVQFBGHJXYZWK'
ETAOIN_EN = 'ETAOINSHRDLCUMWFGYPBVKJXQZ'
LETTERS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

def compute_letter_frequencies(text):
    freq = {}
    total = 0
    for char in text.upper():
        if char.isalpha():
            freq[char] = freq.get(char, 0) + 1
            total += 1
    for letter in freq:
        freq[letter] = (freq[letter] / total) * 100 if total > 0 else 0
    return freq

def sort_by_frequency(freq):
    return ''.join(sorted(freq, key=lambda x: freq[x], reverse=True))

def find_caesar_key_by_frequency(text, language="Français"):
    freq = compute_letter_frequencies(text)
    sorted_letters = sort_by_frequency(freq)
    if not sorted_letters:
        return 0
    most_frequent = sorted_letters[0]
    return (ord(most_frequent) - ord('E')) % 26

def decrypt_by_frequency_analysis(text):
    lang, ic = detect_language(text)
    freq = compute_letter_frequencies(text)
    sorted_letters = sort_by_frequency(freq)

    print(f"  Indice : {ic:.4f} | Langue probable : {lang}")
    print(f"  Lettres par fréquence : {sorted_letters[:10]}...")
    print(f"  Top 5 : ", end="")
    for letter in sorted_letters[:5]:
        print(f"{letter}={freq[letter]:.1f}%", end=" ")
    print()

    key = find_caesar_key_by_frequency(text, lang)
    decrypted = caesar_decode(text, key)
    return decrypted, key, lang

print("=" * 60)
print("Exercice 3.1 - Déchiffrement par analyse fréquentielle")
print("=" * 60)

print("\n--- Message 4 ---")
result4, key4_freq, lang4 = decrypt_by_frequency_analysis(message4)
print(f"  Clé trouvée : {key4_freq}")
print(f"  Texte déchiffré :\n{result4}")

print("\n--- Message 5 ---")
result5, key5_freq, lang5 = decrypt_by_frequency_analysis(message5)
print(f"  Clé trouvée : {key5_freq}")
print(f"  Texte déchiffré :\n{result5}")
```

#### 3.3 - Analyse par digrammes et trigrammes

```python
DIGRAMS_FR = ['ES', 'DE', 'LE', 'EN', 'RE', 'NT', 'ON', 'ER', 'TE', 'EL',
              'AN', 'SE', 'ET', 'LA', 'AI', 'IT', 'ME', 'OU', 'EM', 'IE']

TRIGRAMS_FR = ['ENT', 'LES', 'EDE', 'DES', 'QUE', 'AIT', 'LLE', 'SDE', 'ION', 'EME',
               'ELA', 'RES', 'MEN', 'ESE', 'DEL', 'ANT', 'TIO', 'PAR', 'ESD', 'TDE']

DIGRAMS_EN = ['TH', 'HE', 'IN', 'ER', 'AN', 'RE', 'ON', 'AT', 'EN', 'ND',
              'TI', 'ES', 'OR', 'TE', 'OF', 'ED', 'IS', 'IT', 'AL', 'AR']

TRIGRAMS_EN = ['THE', 'AND', 'ING', 'ION', 'TIO', 'ENT', 'ERE', 'HER', 'ATE', 'VER',
               'TER', 'THA', 'HAT', 'ERS', 'HIS', 'RES', 'ILL', 'ARE', 'CON', 'NCE']

def count_digrams(text):
    clean = ''.join(c.upper() for c in text if c.isalpha())
    digrams = {}
    for i in range(len(clean) - 1):
        d = clean[i:i+2]
        digrams[d] = digrams.get(d, 0) + 1
    return digrams

def count_trigrams(text):
    clean = ''.join(c.upper() for c in text if c.isalpha())
    trigrams = {}
    for i in range(len(clean) - 2):
        t = clean[i:i+3]
        trigrams[t] = trigrams.get(t, 0) + 1
    return trigrams

def top_ngrams(ngrams, n=10):
    return sorted(ngrams.items(), key=lambda x: x[1], reverse=True)[:n]

def match_score(ngrams_text, ref_list):
    total = sum(ngrams_text.values())
    if total == 0:
        return 0.0
    score = 0
    for ng in ref_list:
        if ng in ngrams_text:
            weight = len(ref_list) - ref_list.index(ng)
            score += (ngrams_text[ng] / total) * weight
    return score

def is_clear_text(text):
    digrams = count_digrams(text)
    trigrams = count_trigrams(text)
    score_fr = match_score(digrams, DIGRAMS_FR) + match_score(trigrams, TRIGRAMS_FR)
    score_en = match_score(digrams, DIGRAMS_EN) + match_score(trigrams, TRIGRAMS_EN)
    max_score = max(score_fr, score_en)
    if max_score > 0.015:
        return True, "Français" if score_fr > score_en else "Anglais", score_fr, score_en
    return False, "Indéterminé", score_fr, score_en

def analyze_ngrams(text, name=""):
    print(f"\n{'=' * 60}")
    print(f"Analyse digrammes/trigrammes - {name}")
    print('=' * 60)

    digrams = count_digrams(text)
    trigrams = count_trigrams(text)

    print(f"\nTop 10 digrammes :")
    for d, count in top_ngrams(digrams):
        mark = "*" if d in DIGRAMS_FR or d in DIGRAMS_EN else " "
        print(f"  {mark} {d}: {count}")

    print(f"\nTop 10 trigrammes :")
    for t, count in top_ngrams(trigrams):
        mark = "*" if t in TRIGRAMS_FR or t in TRIGRAMS_EN else " "
        print(f"  {mark} {t}: {count}")

    clear, lang, score_fr, score_en = is_clear_text(text)
    print(f"\nScores - FR: {score_fr:.4f} | EN: {score_en:.4f}")

    if clear:
        print(f"VERDICT : Le message est EN CLAIR ({lang})")
    else:
        print(f"VERDICT : Le message est CHIFFRE")

analyze_ngrams(message4, "Message 4 (chiffré)")
analyze_ngrams(caesar_decode(message4, 3), "Message 4 (déchiffré)")

analyze_ngrams(message6, "Message 6 (chiffré)")
analyze_ngrams(caesar_decode(message6, 7), "Message 6 (déchiffré)")
```

#### 3.4 BONUS - Lettres répétées

```python
def count_repeated_letters(text):
    clean = ''.join(c.upper() for c in text if c.isalpha())
    doubles = {}
    triples = {}

    for i in range(len(clean) - 1):
        pair = clean[i:i+2]
        if pair[0] == pair[1]:
            doubles[pair] = doubles.get(pair, 0) + 1

    for i in range(len(clean) - 2):
        trio = clean[i:i+3]
        if trio[0] == trio[1] == trio[2]:
            triples[trio] = triples.get(trio, 0) + 1

    return doubles, triples

def display_repeated_letters(text, name=""):
    print(f"\n{'=' * 60}")
    print(f"Lettres répétées - {name}")
    print('=' * 60)
    doubles, triples = count_repeated_letters(text)

    print("\nDoublons (2 lettres identiques consécutives) :")
    for pair, count in sorted(doubles.items(), key=lambda x: x[1], reverse=True):
        print(f"  {pair}: {count}")

    if triples:
        print("\nTriplons (3 lettres identiques consécutives) :")
        for trio, count in sorted(triples.items(), key=lambda x: x[1], reverse=True):
            print(f"  {trio}: {count}")
    else:
        print("\nAucun triplon trouvé.")

display_repeated_letters(caesar_decode(message3, 18), "Message 3 (déchiffré - FR)")
display_repeated_letters(caesar_decode(message4, 3), "Message 4 (déchiffré - FR)")
display_repeated_letters(caesar_decode(message6, 7), "Message 6 (déchiffré - EN)")
```
